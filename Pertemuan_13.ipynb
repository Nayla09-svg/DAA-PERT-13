```python
from collections import deque

# BFS
def bfs(graph, start):
    visited = set()
    queue = deque([start])
    urutan = []

    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.add(node)
            urutan.append(node)
            queue.extend(graph[node])
    return urutan

# DFS
def dfs(graph, start):
    visited = set()
    stack = [start]
    urutan = []

    while stack:
        node = stack.pop()
        if node not in visited:
            visited.add(node)
            urutan.append(node)
            stack.extend(reversed(graph[node]))
    return urutan


# DATA GRAPH
graph_amin = {
    "Amin": ["Budi", "Cici"],
    "Budi": ["Deni"],
    "Cici": ["Eka"],
    "Deni": [],
    "Eka": []
}

graph_rektor = {
    "Rektor": ["Dekan1", "Dekan2"],
    "Dekan1": ["KaprodiA", "KaprodiB"],
    "Dekan2": ["KaprodiC"],
    "KaprodiA": [],
    "KaprodiB": [],
    "KaprodiC": []
}

graph_titik0 = {
    0: [1, 2],
    1: [3],
    2: [4, 5],
    3: [],
    4: [],
    5: []
}

# EKSEKUSI
print("BFS - Graph Amin:", bfs(graph_amin, "Amin"))
print("DFS - Graph Amin:", dfs(graph_amin, "Amin"))

print("BFS - Graph Rektor:", bfs(graph_rektor, "Rektor"))
print("DFS - Graph Rektor:", dfs(graph_rektor, "Rektor"))

print("BFS - Graph Titik 0:", bfs(graph_titik0, 0))
print("DFS - Graph Titik 0:", dfs(graph_titik0, 0))
```
